```python
import random as ran
```

```python
def H(r, n):
    out = ""
    ran.seed(r)
    for i in range(n):
        out = out + chr(ran.randint(0, 128))
    return out
```

```python
#A working example of how that so-called Hash function works
H(1, n), len(H(1, n))
```

```python
def dec2bin(n, l):
  out = ""
  m = n
  while m>0:
    out = out + str(m%2)
    m = int(m/2)
  #print(len(out))
  if l>len(out):
    out = (l-len(out))*"0"+out
  else: pass
  return out
dec2bin(3, 9)
```

```python
def QFT(state):
  n = len(state.dims())
  qc = qk.QuantumCircuit(n)
  qc.initialize(state)
  for i in range(n):
    qc.h(i)
    for j in range(i, n-1): #(0,n-i-1):
      qc.unitary(np.array([[1, 0], [0, np.exp(2j*np.pi/(2**(j-i+1)))]], dtype=complex), [j+1, i])
      #qc.crx(np.pi/2, j+1, i) #n-j-1, i)
  return qc
```
